````python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by='Date').reset_index(drop=True)
print("DataFrame after converting 'Date' to datetime and sorting:")
df.head()
````

````python
df['flow_lag_1d'] = df['Daily flow Rate'].shift(1)
df['flow_lag_3d'] = df['Daily flow Rate'].shift(3)
print("DataFrame with new 'flow_lag_1d' and 'flow_lag_3d' features:")
df.head()
````

````python
df['flow_change_rate'] = df['Daily flow Rate'] - df['flow_lag_1d']
print("DataFrame with new 'flow_change_rate' feature:")
df.head()
````

````python
decay_factor = 0.85
df['API'] = 0.0

# Initialize the first API value
df.loc[0, 'API'] = df.loc[0, '24h accumulated']

# Calculate API for subsequent days
for i in range(1, len(df)):
    df.loc[i, 'API'] = (df.loc[i-1, 'API'] * decay_factor) + df.loc[i, '24h accumulated']

print("DataFrame with new 'API' feature:")
df.head()
````

````python
df['24h_accumulated_roll_3d'] = df['24h accumulated'].rolling(window=3).sum()
df['24h_accumulated_roll_5d'] = df['24h accumulated'].rolling(window=5).sum()
df['24h_accumulated_roll_7d'] = df['24h accumulated'].rolling(window=7).sum()
print("DataFrame with new rolling sum features for '24h accumulated':")
df.head(10)
````

````python
df['saturation_index'] = (df['sm_top_0_10cm'] + df['sm_10_40cm']) / 2
print("DataFrame with new 'saturation_index' feature:")
df.head()
````

````python
print("DataFrame with all engineered features:")
df.head(10)
````

````python
df['soil_moisture_change_rate'] = df['saturation_index'].diff()
print("DataFrame with new 'soil_moisture_change_rate' feature:")
df.head()
````

## Normalize Features

### Subtask:
Scale all relevant numerical features to a common range (e.g., 0 to 1) to ensure that features with larger magnitudes do not disproportionately influence the flood risk score.


**Reasoning**:
First, I'll identify all numerical columns in the DataFrame `df` that need to be scaled, excluding the 'Date' column.



````python
numerical_cols = df.select_dtypes(include=['number']).columns.tolist()
# Exclude 'Date' column if it's in the numerical columns list (though it should be datetime now)
if 'Date' in numerical_cols:
    numerical_cols.remove('Date')

print(f"Numerical columns identified for scaling: {numerical_cols}")
````

**Reasoning**:
Now that the numerical columns have been identified, I will import `MinMaxScaler`, initialize it, and then apply it to scale these numerical columns to a range between 0 and 1. The scaled values will replace the original numerical columns in the DataFrame.



````python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

print("DataFrame after scaling numerical features using MinMaxScaler:")
df.head()
````

## Determine Feature Importance (Weights)

### Subtask:
Calculate the absolute Pearson correlation of each normalized feature with the 'Daily flow Rate'. These correlation values will serve as weights, representing each feature's importance in relation to a primary flood indicator.


**Reasoning**:
To calculate the Pearson correlation of each numerical feature with 'Daily flow Rate', I first need to ensure that 'Daily flow Rate' is not correlated with itself and that the 'Date' column is excluded, as specified in the instructions. Then, I will compute the absolute correlations.



````python
numerical_cols_for_correlation = [col for col in numerical_cols if col not in ['Date', 'Daily flow Rate']]

# Calculate Pearson correlation of each feature with 'Daily flow Rate'
correlations = df[numerical_cols_for_correlation].corrwith(df['Daily flow Rate'])

# Take the absolute value of the correlation coefficients
feature_weights = correlations.abs()

print("Feature Importance Weights (Absolute Pearson Correlation with 'Daily flow Rate'):")
print(feature_weights)
````

## Calculate Weighted Flood Risk Score

### Subtask:
Compute a composite 'Weighted Flood Risk Score' for each day by summing the product of each normalized feature and its corresponding importance weight.


**Reasoning**:
As instructed, I will initialize a new column 'Weighted Flood Risk Score' in the DataFrame `df` with zeros. This column will accumulate the weighted sum of each normalized feature.



````python
df['Weighted Flood Risk Score'] = 0.0
print("DataFrame with initialized 'Weighted Flood Risk Score' column:")
df.head()
````

**Reasoning**:
Now that the 'Weighted Flood Risk Score' column is initialized, I will iterate through the numerical features (excluding 'Daily flow Rate'), multiply each feature's normalized values by its corresponding absolute Pearson correlation weight, and sum these products into the 'Weighted Flood Risk Score' column. Finally, I will display the first few rows to verify the calculation.



````python
for feature in numerical_cols_for_correlation:
    df['Weighted Flood Risk Score'] += df[feature] * feature_weights[feature]

print("DataFrame after calculating 'Weighted Flood Risk Score':")
df.head()
````

## Assign Flood Risk Labels

### Subtask:
Discretize the 'Weighted Flood Risk Score' into four distinct classes (0, 1, 2, 3) to represent different levels of flood risk. This can be done using quantiles to create balanced classes or by defining specific thresholds.


**Reasoning**:
First, I will define the number of classes for the flood risk score and then use `pd.qcut()` to discretize the 'Weighted Flood Risk Score' into these classes, assigning the resulting labels to a new column 'Flood Risk Label'.



````python
num_classes = 4
df['Flood Risk Label'] = pd.qcut(df['Weighted Flood Risk Score'], q=num_classes, labels=False, duplicates='drop')

print(f"Value counts for 'Flood Risk Label' with {num_classes} classes:")
print(df['Flood Risk Label'].value_counts().sort_index())

print("\nFirst 5 rows of the DataFrame with 'Flood Risk Label':")
df.head()
````
