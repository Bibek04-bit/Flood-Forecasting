## Prepare DataFrame for Time-Series Analysis

### Subtask:
Convert the 'Date' column to datetime objects and sort the DataFrame by date to ensure correct time-series calculations.


**Reasoning**:
Convert the 'Date' column to datetime objects and sort the DataFrame by date for time-series analysis.



```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by='Date').reset_index(drop=True)
print("DataFrame after converting 'Date' to datetime and sorting:")
df.head()
```

## Calculate Flow Lag Features

### Subtask:
Calculate 'flow_lag_1d' and 'flow_lag_3d' by shifting the 'Daily flow Rate' column by 1 and 3 days, respectively.


**Reasoning**:
To calculate 'flow_lag_1d' and 'flow_lag_3d', I will shift the 'Daily flow Rate' column by 1 and 3 periods, respectively, using the `.shift()` method, and then display the head of the DataFrame to confirm the new columns.



```python
df['flow_lag_1d'] = df['Daily flow Rate'].shift(1)
df['flow_lag_3d'] = df['Daily flow Rate'].shift(3)
print("DataFrame with new 'flow_lag_1d' and 'flow_lag_3d' features:")
df.head()
```

## Calculate Flow Change Rate

### Subtask:
Calculate the 'flow change rate' by computing the difference between the current 'Daily flow Rate' and the 'Daily flow Rate' from the previous day.


**Reasoning**:
To calculate the 'flow change rate', I will subtract the 'flow_lag_1d' from the 'Daily flow Rate' and store it in a new column, then display the head of the DataFrame to show the new feature.



```python
df['flow_change_rate'] = df['Daily flow Rate'] - df['flow_lag_1d']
print("DataFrame with new 'flow_change_rate' feature:")
df.head()
```

## Calculate Antecedent Precipitation Index (API)

### Subtask:
Calculate a simple Antecedent Precipitation Index (API) using the '24h accumulated' precipitation data, assuming a decay factor for previous days.


**Reasoning**:
To calculate the Antecedent Precipitation Index (API), I will define a decay factor and then iterate through the DataFrame to apply the API formula, which depends on the previous day's API value and the current day's accumulated precipitation. I will then display the head of the DataFrame to show the new 'API' feature.



```python
decay_factor = 0.85
df['API'] = 0.0

# Initialize the first API value
df.loc[0, 'API'] = df.loc[0, '24h accumulated']

# Calculate API for subsequent days
for i in range(1, len(df)):
    df.loc[i, 'API'] = (df.loc[i-1, 'API'] * decay_factor) + df.loc[i, '24h accumulated']

print("DataFrame with new 'API' feature:")
df.head()
```

## Calculate Rolling Sums for Precipitation

### Subtask:
Calculate 3-day, 5-day, and 7-day rolling sums for '24h accumulated' precipitation.

**Reasoning**:
To calculate the rolling sums for '24h accumulated' precipitation, I will use the `.rolling()` method on the '24h accumulated' column with window sizes of 3, 5, and 7 days, respectively, and then display the head of the DataFrame to confirm the new features.



```python
df['24h_accumulated_roll_3d'] = df['24h accumulated'].rolling(window=3).sum()
df['24h_accumulated_roll_5d'] = df['24h accumulated'].rolling(window=5).sum()
df['24h_accumulated_roll_7d'] = df['24h accumulated'].rolling(window=7).sum()
print("DataFrame with new rolling sum features for '24h accumulated':")
df.head(10)
```

## Create Saturation Index

### Subtask:
Create a 'saturation index' from 'sm_top_0_10cm' and 'sm_10_40cm' by averaging these two soil moisture measurements.

**Reasoning**:
To calculate the 'saturation index', I will average the 'sm_top_0_10cm' and 'sm_10_40cm' columns and store it in a new column, then display the head of the DataFrame to show the new feature.



```python
df['saturation_index'] = (df['sm_top_0_10cm'] + df['sm_10_40cm']) / 2
print("DataFrame with new 'saturation_index' feature:")
df.head()
```
